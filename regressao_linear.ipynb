```
#importando a biblioteca para visualização de dados
import matplotlib.pyplot as plt
import numpy as np
```

```
#dados
x = [-1.,-0.7, -0.5, -0.3, -0.1, 0.1, 0.3, 0.5, 0.7, 1.]
y = [-1.13, -0.57, -0.22, 0.54, 0.49, 1.15, 1.64, 2.17, 2.64, 2.95]
```

```
#plot dos dados
plt.figure(figsize=(10,5))
plt.plot(x, y,'o', label='dados originais')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```
#iremos estimar uma função tipo y=ax+b
#transformando x e y em arrays e vetor coluna
x, y = np.array(x).reshape(-1,1), np.array(y).reshape(-1,1)
```

```
#adicionando as bias: para estimar o b
X = np.hstack((x, np.ones(x.shape)))
```

```
#estimando a e b
beta = np.linalg.pinv(X).dot(y)
print('a estimado: ', beta[0][0])
print('b estimado: ', beta[1][0])
```

```
#plotando os dados
plt.figure(figsize=(10,5))
plt.plot(x, y, 'o', label='dados originais')
plt.plot(x, X.dot(beta), label='regressão linear')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regressão linear com numpy')
plt.grid()
plt.show()
```
